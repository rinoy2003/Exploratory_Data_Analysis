# BANK TRANSACTION

## INTRODUCTION

The bank transactions dataset contains detailed information about individual transactions made by customers. Each record includes attributes such as TransactionID, Account ID, and Transaction Amount, along with the Transaction Date and Transaction Type (e.g., withdrawal or deposit). It also includes Location, Device ID, and IP Address to track the origin of each transaction, as well as Merchant ID and Channel (e.g., online or in-branch). Demographic information such as Customer Age and Occupation is included, along with Transaction Duration, Login Attempts, Account Balance, and the Previous Transaction Date. This dataset is valuable for analyzing customer behavior, detecting fraud, and understanding transaction trends.

### QUESTIONS AND ANSWERS


1. WHAT IS THE TOTAL ACCOUNT BALANCE ACROSS ALL ACCOUNTS?

```python
total_balance = df['Account_Balance'].sum()
print(total_balance)


plt.figure(figsize=(6, 6))
plt.bar(['Total Account Balance'], [total_balance], color='skyblue')
plt.title('Total Account Balance Across All Accounts', fontsize=14)
plt.ylabel('Account Balance', fontsize=12)
plt.xlabel('Total', fontsize=12)
plt.show()
```

2. HOW MANY TRANSACTION OCCURED ON A SPECIFIC DATE

```python

df['Transaction_Date'] = pd.to_datetime(df['Transaction_Date'])
transactions_per_date = df.groupby('Transaction_Date').size()
print(transactions_per_date)

plt.figure(figsize=(8, 6))
transactions_per_date.plot(kind='bar', color='lightgreen')
plt.title('Number of Transactions Per Date')
plt.xlabel('Transaction Date')
plt.ylabel('Number of Transactions')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```


3. WHAT IS THE AVERAGE TRANSACTION AMOUNT FOR EACH TRANSACTION TYPE?

```python
avg_transaction_amount = df.groupby('Transaction_Type')['Transaction_Amount'].mean()
print(avg_transaction_amount)

plt.figure(figsize=(8, 6))
avg_transaction_amount.plot(kind='bar', color='pink')
plt.title('Average Transaction Amount for Each Transaction Type')
plt.xlabel('Transaction Type')
plt.ylabel('Average Transaction Amount')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```


4. WHAT IS THE MOST COMMON TRANSACTION TYPE FOR CUSTOMERS OF A SPECIFIC AGE GROPUP(CUSTOMER AGE)?

```python

specific_age_group = 25 

age_group_data = df[df['Customer_Age'] == specific_age_group]
transaction_counts = age_group_data['Transaction_Type'].value_counts()
print(transaction_counts)


plt.figure(figsize=(8, 6))
plt.pie(transaction_counts, labels=transaction_counts.index, autopct='%1.1f%%', colors=['skyblue','lightgreen'], startangle=90)
# plt.title(f'Most Common Transaction Types for Customers Aged {age_group_data}')
plt.axis('equal')  
plt.tight_layout()
plt.show()

```

5. HOW MANY TRANSACTIONS WERE MADE FROM EACH LOCATION?

```python
transactions_by_location = df['Location'].value_counts()
print(transactions_by_location)

plt.figure(figsize=(8, 6))
transactions_by_location.plot(kind='bar', color='orange')

plt.title('Number of Transactions by Location')
plt.xlabel('Location')
plt.ylabel('Number of Transactions')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

6. WHAT IS THE AVERAGE TRANSACTION DURATION ACROSS ALL TRANSACTION?

```python

average_duration = df['Transaction_Duration'].mean()
print(average_duration)


plt.figure(figsize=(6, 6))
plt.bar('Average Duration', average_duration, color='lightgreen')
plt.title('Average Transaction Duration Across All Transactions')
plt.ylabel('Transaction Duration (seconds)')
plt.tight_layout()
plt.show()

# print(f"Average Transaction Duration: {average_duration} seconds")
```


7. WHAT IS THE AVERAGE TRANSACTION AMOUNT PER MERCHANT ACROSS ALL CUSTOMERS?

```python
avg_amount_per_merchant = df.groupby('Merchant_ID')['Transaction_Amount'].mean()
print(avg_amount_per_merchant)

plt.figure(figsize=(12, 6))
avg_amount_per_merchant.plot(kind='bar', color='lightgreen', edgecolor='black')
plt.title('Average Transaction Amount Per Merchant')
plt.xlabel('Merchant ID')
plt.ylabel('Average Transaction Amount')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability if needed
plt.grid(True)
plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()
```


8. WHAT IS THE CORRELATION BETWEEN THE ACCOUNT BALANCE AND TRANSACTION AMOUNT?

```python

correlation = df['Account_Balance'].corr(df['Transaction_Amount'])
print(correlation)

# print(f"Correlation between AccountBalance and TransactionAmount: {correlation}")

plt.figure(figsize=(8, 6))
plt.scatter(df['Account_Balance'], df['Transaction_Amount'], color='red')
plt.title('Relationship Between Account Balance and Transaction Amount')
plt.xlabel('Account Balance')
plt.ylabel('Transaction Amount')
plt.tight_layout()
plt.show()
```


9. HOW MUCH TIME (TransactionDuration) DOES THE AVERAGE CUSTOMER TAKE TO COMPLETE A TRANSACTION BASED ON TRANSACTION TYPE?

```python

avg_duration_by_type = df.groupby('Transaction_Type')['Transaction_Duration'].mean()
print(avg_duration_by_type)


plt.figure(figsize=(8, 6))
avg_duration_by_type.plot(kind='bar', color='skyblue')
plt.title('Average Transaction Duration by Transaction Type')
plt.xlabel('Transaction Type')
plt.ylabel('Average Transaction Duration (seconds)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()



```


10. WHAT IS THE AVERAGE AGE(CUSTOMER AGE) OF CUSTOMERS WHO MAKE TRANSACTION OF CERTAIN TYPE(transaction type)?

```python

avg_age_by_transaction_type = df.groupby('Transaction_Type')['Customer_Age'].mean()
print(avg_age_by_transaction_type)


plt.figure(figsize=(8, 6))
avg_age_by_transaction_type.plot(kind='bar', color='lightgreen')
plt.title('Average Age of Customers by Transaction Type')
plt.xlabel('Transaction Type')
plt.ylabel('Average Age')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()


```

11. HOW DO TRANSACTION AMOUNTS COMPARE ACROSS DIFFERENT LOCATIONS?

```python

transaction_sum_by_location = df.groupby('Location')['Transaction_Amount'].sum()
print(transaction_sum_by_location)


plt.figure(figsize=(8, 6))
transaction_sum_by_location.plot(kind='bar', color='brown')
plt.title('Transaction Amounts by Location')
plt.xlabel('Location')
plt.ylabel('Total Transaction Amount')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()


```


12. WHAT IS THE DISTRIBUTION OF TRANSACTION TYPES ACROSS DIFFERENT CHANNELS?

```python

transaction_distribution = df.groupby(['Transaction_Type','Channel']).size().unstack()
print(transaction_distribution)


transaction_distribution.plot(kind='bar',stacked=True, figsize=(10, 6))
plt.title('Distribution of Transaction Types Across Different Channels')
plt.xlabel('Transaction Type')
plt.ylabel('Number of Transactions')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()  
plt.show()
```


13. WHAT IS THE FREQUENCY OF LOGIN ATTEMPTS OVER TIME FOR CUSTOMERS?

```python

df['transaction_date'] = pd.to_datetime(df['Transaction_Date'])
df.set_index('transaction_date', inplace=True)
login_attempts_per_day = df.resample('D')['Login_Attempts'].sum() 
print(login_attempts_per_day)


plt.figure(figsize=(10, 6))
login_attempts_per_day.plot(kind='line', marker='o', color='b')
plt.title('Frequency of Login Attempts Over Time', fontsize=14)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Login Attempts', fontsize=12)
plt.xticks(rotation=45)  
plt.grid(True)
plt.show()
```



14. HOW DO TRANSACTION AMOUNTS DIFFER BY CUSTOMER OCCUPATION?

```python

occupation_transaction_amount = df.groupby('Customer_Occupation')['Transaction_Amount'].mean().sort_values(ascending=False)
print(occupation_transaction_amount)

plt.figure(figsize=(8, 8))
occupation_transaction_amount.plot(kind='pie', autopct='%1.1f%%', startangle=90, cmap='Set3')
plt.title('Transaction Amount Distribution by Customer Occupation', fontsize=16)
plt.ylabel('') 
plt.tight_layout()
plt.show()
```
